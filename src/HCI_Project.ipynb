```python
import os
files=os.listdir("C://Users/PrishitaRay/Desktop/DSER_Movie_Subtitles/Switchboard Dialog Act Corpus/swda")
sw_train=[]

for i in range(1,len(files)):
        sw_train.append(files[i])
        
print(len(sw_train))
```

```python
#Training on SwitchBoard Dialogs Act Corpus
import re
os.chdir("C://Users/PrishitaRay/Desktop/DSER_Movie_Subtitles/Switchboard Dialog Act Corpus/swda")
tokenized_dialogs=[]
dialog_acts=[]
```

```python
dialogs=[]
for f in sw_train:
    file= pd.read_csv(f)
    d_act= list(file['act_tag'])
    text_with_pos= list(file['pos'])
    
    for d in d_act:
        dialog_acts.append(d)

    for i in range(0,len(d_act)):
        #Removing unneccessary delimiters from text
        pos= str(text_with_pos[i]).strip()
        word_tag = map((lambda x : tuple(x.split("/"))), re.split(r"\s+", pos))
        word_tag = filter((lambda x : len(x) == 2), word_tag)

        w= [x[0] for x in word_tag]
        tokenized_dialogs.append(w)
        separator=' '
        d= separator.join(w)
        dialogs.append(d)
```

```python
print(len(dialogs))
print(len(dialog_acts))
print(dialogs[0:5])
print(dialog_acts[0:5])
```

```python
#Testing for first movie only
test_data=[]
for key in pp_lines_dict.keys():
    #List of sentences with word embeddings for each movie
    swe=[]
    for line in pp_lines_dict[key]:
        tokenized = re.sub('[,?.]','', line).lower().split(' ')
        separator=' '
        line=separator.join(tokenized)
        test_data.append(line)
        
    break
```

```python
#List of words
low=[]

for dialog in tokenized_dialogs:
    for token in dialog:
        low.append(token)  

low_cnt={k:v for v,k in enumerate(np.unique(low))}

sorted_lc=sorted(low_cnt.items(), key = lambda kv:(kv[1], kv[0]),reverse=True)[0:10000]
vocab=dict(sorted_lc)
print(len(vocab))
print(vocab)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from keras.preprocessing.text import one_hot
from keras.preprocessing.sequence import pad_sequences
VOCAB_LEN=len(vocab.keys())
n_timesteps=20
k=5
encoded_docs = [one_hot(d, VOCAB_LEN) for d in dialogs]
max_length = 20
padded_docs = pad_sequences(encoded_docs, maxlen=max_length, padding='post')

X= padded_docs[0:100000]
dialog_acts=np.array(dialog_acts)
label_encoder = LabelEncoder()
vec = label_encoder.fit_transform(dialog_acts)
enc= OneHotEncoder(sparse=False)
Y = enc.fit_transform(vec.reshape((vec.shape[0]),1))
Y= Y[0:100000]
```

```python
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Flatten
from keras.layers import Bidirectional
from keras.layers import TimeDistributed
from keras.layers import Dense
from keras import optimizers
from keras.layers.embeddings import Embedding

model= Sequential()
model.add(Embedding(VOCAB_LEN, 200, input_length=20))
model.add(Dropout(0.5))
model.add(Bidirectional(LSTM(100,return_sequences=True)))
#k=5 for previous 5 timesteps
model.add(LSTM(100))
#One hot encoding of Dialog Act class
model.add(Dense(303, activation='softmax'))
sgd = optimizers.SGD(lr=1, decay=0.8, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
print(model.summary())
model.fit(X, Y, epochs=50,batch_size=32,verbose=2)
# evaluate the model
loss, accuracy = model.evaluate(X, Y, verbose=2)
print('Accuracy: %f' % (accuracy*100))
```

```python
encoded_docs = [one_hot(d, VOCAB_LEN) for d in test_data]
max_length = 20
X_test = pad_sequences(encoded_docs, maxlen=max_length, padding='post')
ohe_Y_pred=model.predict(X_test)
```

```python
#Rules to determine speaker turns based on rules for a pair of dialog acts:
turn_indexes=[]
for i in range(0,len(pred_DA)-1):
    #d is for the current dialog, nd is for the next dialog in the pair
    d=pred_DA[i]
    nd=pred_DA[i+1]
    if nd=='b' or nd=='sv' or nd=='aa' or nd=='ba' or nd=='bk' or nd=='fo_o_fw_by_bc' or nd=='bf' or nd=='ft' or nd=='bd':
        turn_indexes.append(i+1)
    elif d=='qo' or d=='qy^d' or d=='qy' or d=='qw' or d=='^g' or d=='qrr':
        if nd=='ny' or nd=='nn' or nd=='na' or nd=='ng' or nd=='no' or nd=='sd' or nd=='sv'or nd=='arp_nd' or nd=='aap_am' or nd=='ng' or nd=='oo_co_cc' or nd=='ar^r':
            turn_indexes.append(i+1)
    elif d=='ad' and nd=='no':
        turn_indexes.append(i+1)
    elif d=='fp' and nd=='sd':
        turn_indexes.append(i+1)
    elif d=='qw^d':
        if nd=='sd' or nd=='ng':
            turn_indexes.append(i+1)
    elif d=='fc' and nd=='aa':
        turn_indexes.append(i+1)
    elif d=='%' or d=='ba' or d=='sv' or d=='sd':
        if nd=='^q':
            turn_indexes.append(i+1)
    elif d=='qh':
        turn_indexes.append(i+1)
    elif d=='fa' or d=='ft' or d=='oo_co_cc' or d=='fc':
        if nd=='bd':
            turn_indexes.append(i+1)
    elif d=='bf' and nd=='aap_am':
        turn_indexes.append(i+1)
    elif d=='bh' or d=='^g':
        if nd=='aa':
            turn_indexes.append(i+1)
    elif d=='t3' and nd=='x':
        turn_indexes.append(i+1)
    elif d=='^q' and nd=='br':
        turn_indexes.append(i+1)
    elif d=='^h':
        if nd=='qw^d' or nd=='br' or nd=='bd' or nd=='^q':
            turn_indexes.append(i+1)
    elif d=='br' and nd=='ad':
        turn_indexes.append(i+1)
    elif d=='arp_nd':
        if nd=='t3' or nd=='oo_co_cc' or nd=='bd' or nd=='ft' or nd=='qw^d':
            turn_indexes.append(i+1)
    elif d=='^2':
        turn_indexes.append(i+1)
    elif d=='h':
        if nd=='bd' or nd=='^q':
            turn_indexes.append(i+1)
    elif d=='t1':
        if nd=='qw^d' or nd=='t3' or nd=='arp_nd' or nd=='no' or nd=='b^m':
            turn_indexes.append(i+1)
    elif d=='ad' and nd=='ft':
        turn_indexes.append(i+1)
    elif d=='%':
        turn_indexes.append(i+1)
    elif d=='fo_o_fw_by_bc' and nd=='bk':
        turn_indexes.append(i+1)
```
